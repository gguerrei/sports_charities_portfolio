```python
#Now for the analysis on the overall updated data. My main goal here is to see how many charities spent more on other expenses than on donations.
```

```python
#to do that, I followed the same structure I used for my original analysis, considering that true/false values is probably the quickest way to filter the data into a dataframe
#for sake of convenience, I broke it down in the following cell:
```

```python
df_updated = pd.read_csv('data/sports_charities_updated_portfolio.csv')
#brings the new updated team csv into the analysis
df_updated_filtered = df_updated['Other Expenses (I - 17)'] - df_updated['Functional Expenses - Grants and other assistance to domestic organizations and domestic governments, Total Expenses, Column A (Part IX1)'] > 0
#returns true/false values for whether charities spent more on other expenses than on donations
teams_more_expenses = df_updated_filtered.to_frame()
#turns the true/false value from a series into a dataframe called teams_more_expenses
teams_more_expenses_true = teams_more_expenses[teams_more_expenses[0] == True]
#filters out for only the teams listed as "true", which means they spent more on other expenses than on donations
teams_more_expenses_true
#checking to see if it worked
```

```python
#now, I'm trying to add more information to this dataframe, which makes understanding it easier. For that, I'll add some columns from the updated csv file.
#to do that, I'll follow the same steps I made previously.
```

```python
team_names_column_updated = df_updated['Team name']
#created a variable for the team names column
other_expenses_column = df_updated['Other Expenses (I - 17)']
#created a variable for the other expenses column
donations_column = df_updated['Functional Expenses - Grants and other assistance to domestic organizations and domestic governments, Total Expenses, Column A (Part IX1)']
#created a variable for the donations column
tax_year_column_updated = df_updated['Tax year (the year in the top right corner)']
#created a variable for the tax year column
teams_more_expenses_true.insert(0, 'Team name', team_names_column_updated)
teams_more_expenses_true.insert(1,'Year', tax_year_column_updated)
teams_more_expenses_true.insert(2, 'Other Expenses', other_expenses_column)
teams_more_expenses_true.insert(3, 'Donations', donations_column)
#added all of the columns to the dataframe
teams_more_expenses_true
#checking to see if it worked
```

```python
len(teams_more_expenses_true)
#checking to see how many teams spent more on other expenses than on donations
```

```python
#while that works, we can also structure it differently so the same team in a different year doesn't account for two different entries.
```

```python
teams_more_expenses_true.groupby('Team name').size()
#this works the same as a regular groupby, but since its a dataframe, we need to add the size command, which essentially just shows us how many times a value appears in the dataframe.
#in this case, its counting how many times a team appeared in the dataframe.
#For our purposes, this removes the duplicates, leaving us with the pure number of teams that spent more on other expenses than on donations, regardless of year. 
```

```python
len(teams_more_expenses_true.groupby('Team name').size())
#counting how many teams there are
```

```python
#20! 
#To check if this is correct, I'll need to make similar calculations in the original excel file outside of python to see if the numbers match.
#In this case, since I also checked on Excel, I can say for sure that my findings here are correct. 
```

```python
#Now that we found that out, we should probably also calculate the difference between both values. 
#As in, how much more did these teams spend on other expenses than on donations?
#To do that, we first must add a column to the dataframe that calculates the difference between both values.
```

```python
teams_more_expenses_true['Difference'] = teams_more_expenses_true['Other Expenses'] - teams_more_expenses_true['Donations']
#since we're making a new column with a number value, doing this the straightforward way is also the easiest. 
teams_more_expenses_true
#checking to see if it worked
```

```python
teams_more_expenses_true.sort_values(by='Difference', ascending=False)
#seeing which teams had the largest difference 
```
